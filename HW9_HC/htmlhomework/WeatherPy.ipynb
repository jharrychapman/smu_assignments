# Analysis
----

* Temperature shows a correlation with latitude whereby temperature increases as it approaches a latitude of zero from both positive and negative latitudes

* Wind speed demonstrates low correlation with latitude from our data

* Wind speed concentrates in the band of 0mph and 10mph regardless of latitude, yet there are still a significant number of cities with wind speeds in the range of 10mph to 20mph with a few cities skewing even higher

* Humidity shows ow correlation wiht latitude from our data

* The bulk of cities show humidities between 60% and above, regardless of latitude

* Cloudiness does not appear to be correlated to latitude from our data

* Of the data in the dataset, only temperature appears to be correlated with latitude with wind speed, humidity, and cloudiness showing low correlation.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read in CSV
weather_df = pd.read_csv('Resources/cities.csv')

# Check counts to make sure even
print(weather_df.count())

weather_df = weather_df.set_index('City_ID')
weather_df.head()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
# Create latitude vs. max temp scatter plot
plt.scatter(weather_df['Lat'], weather_df['Max Temp'], edgecolors='black', facecolors = 'lightblue')
plt.title('City Latitude vs. Max Temperature')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.grid(b = True, which = 'Major', axis = 'both', linestyle = '-', color = 'lightgrey')
plt.savefig("Resources/assets/images/latvtemp.png")
plt.savefig('latvtemp2.png')
plt.show()
```

#### Latitude vs. Humidity Plot

```python
# Create latitude vs humidity scatter plot
plt.scatter(weather_df['Lat'], weather_df['Humidity'], edgecolors='black', facecolors='lightblue')
plt.title('City Latitude vs. Humidity')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.ylim(0,100)
plt.grid(b=True, which='major', axis='both', linestyle='-', color='lightgrey')
plt.savefig('Resources/assets/images/latvhumidity.png')
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
# Create latitude vs. cloudiness scatterplot
plt.scatter(weather_df['Lat'],weather_df['Cloudiness'], edgecolors='black', facecolors='lightblue')
plt.title('City Latitude vs. Cloudiness')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.grid(b = True, which='Major', axis='both', linestyle='-',color='lightgrey')
plt.savefig('Resources/assets/images/latvcloud')
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
# Create latitude vs. wind speed scatter plot
plt.scatter(weather_df['Lat'], weather_df['Wind Speed'], edgecolors='black', facecolors='lightblue')
plt.title('City Latitude vs. Wind Speed')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.grid(b= True, which='major', axis='both', linestyle='-', color='lightgrey')
plt.ylim(-2, 34)
plt.savefig('Resources/assets/images/latvwind.png')
plt.show()
```
